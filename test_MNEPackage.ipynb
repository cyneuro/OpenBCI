```python
import numpy as np

import mne

#MNE can't process .txt files so would have to convert to array first. This is just to test the saved data loading capabilities
sample_data_folder = mne.datasets.sample.data_path() 
sample_data_raw_file = (
    sample_data_folder / "MEG" / "sample" / "sample_audvis_filt-0-40_raw.fif"
)
raw = mne.io.read_raw_fif(sample_data_raw_file)
```

```python
#Print basic details
print(raw)
print(raw.info)

#Print power spectrum density plot
raw.compute_psd(fmax=50).plot(picks="data", exclude="bads", amplitude=False)
raw.plot(duration=5, n_channels=30)

#Can specify parameters for detecting events and make a dictionary+plot displaying them
```
