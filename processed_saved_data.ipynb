```python
import pandas as pd

file_path = 'data/blinking_data.txt'
# Read the data while skipping the comment lines starting with '%'
df = pd.read_csv(file_path, comment='%')

df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

sampling_frequency = 200  

time = np.arange(0, len(df[' EXG Channel 1']) / sampling_frequency, 1 / sampling_frequency)

# Plot the data
plt.plot(time, df[' EXG Channel 0'],label='Channel 1')
plt.plot(time, df[' EXG Channel 1']+500,label='Channel 2')
plt.plot(time, df[' EXG Channel 2']+1000,label='Channel 3')
plt.plot(time, df[' EXG Channel 3']+1500,label='Channel 4')
plt.xlabel('Time (seconds)')  
plt.xlim(40,80)
plt.show()
```

### blinking was 5 seconds on 5 seconds off which can easily be seen in the figure
